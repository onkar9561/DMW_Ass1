```python
# Data Reduction
# Remove/reduce unnecessary attributes from dataframe
subset = df[['movie_title','genres','imdb_score','budget','language',]]
subset.head()
```

```python
# Data Cleaning
# Find missing values
subset.isnull().sum()
```

```python
# Find the missing values from Rating Column and repalce by Median value
# user Defined function
def impute_median(series):
    return series.fillna(series.median())
```

```python
# Fill missing value of integer column with median
subset.budget=df['budget'].transform(impute_median)
```

```python
subset.isnull().sum()
```

```python
# Fill missing value of integer column with mode
subset['language'] = subset['language'].fillna(subset['language'].mode()[0])
```

```python
subset.isnull().sum()
```

```python
# Data Reduction 
subset1 = df[['movie_title', 'genres', 'imdb_score', 'budget']]
subset1.head()
```
